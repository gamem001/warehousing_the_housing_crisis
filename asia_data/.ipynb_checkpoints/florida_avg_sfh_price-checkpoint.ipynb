```python
import pandas as pd
from sqlalchemy import create_engine
from config import password
```

```python
csv = pd.read_csv('single_family_homes.csv')
csv
```

```python
florida = csv.loc[csv['RegionName'] == 'Florida']
florida
```

```python
col_list = ['StateName', '6/30/2006','6/30/2007', '6/30/2008', '6/30/2009', '6/30/2010', '6/30/2011', '6/30/2012', '6/30/2013', '6/30/2014', '6/30/2015', '6/30/2016', '6/30/2017', '6/30/2018', '6/30/2019', '6/30/2020']
florida_homes = florida[col_list]
florida_homes
```

```python
median_fl_sfh = pd.DataFrame ({
    'avg_sale_price':[279318, 269232, 223001, 175795, 163062, 145904, 145116, 161685, 181107, 193642, 210815, 228108, 244794, 255701, 267493],
    'year':[2006, 2007, 2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020]
})

median_fl_sfh
```

```python
postgres_conn_string = f"postgres:{password}" 
engine = create_engine(f"postgresql://{postgres_conn_string}@localhost:5432/housing", paramstyle="format")
```

```python
median_fl_sfh.to_sql(name='housing', con=engine, if_exists='append', index=False)
```
